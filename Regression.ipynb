```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression
from scipy import stats
import statsmodels.api as s


tbl=pd.read_excel(r'D:\DATA ANALYST\data sets\SLR1.XLSX',sheet_name='Sheet3')
tbl.plot('Miles travel','Petrol Consumption in litre',style='o')
plt.ylabel('Petrol Consumption in litre')
plt.title('Miles Traveled with given petrol consumption')
plt.show()

t=tbl['Miles travel']
c=tbl['Petrol Consumption in litre']

t=s.add_constant(t)
model1=sm.OLS(c,t)
result1=model1.fit()
print(result1.summary())


```
